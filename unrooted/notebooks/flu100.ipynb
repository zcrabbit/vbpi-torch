```python
from dataManipulation import *
from utils import summary, summary_raw, get_support_from_mcmc, mcmc_treeprob
from vbpi import VBPI
```

```python
# load the sequence data and estimate the subsplit support
data, taxa = loadData('../data/flu_datasets_fasta/flu100.fasta', 'fasta')
tree_dict_mcmc, tree_names_mcmc, tree_wts_mcmc = mcmc_treeprob('../data/mcmc_flu_data/flu100.trprobs', 'nexus', taxon='keep')
rootsplit_supp_dict, subsplit_supp_dict = get_support_from_mcmc(taxa, tree_dict_mcmc, tree_names_mcmc)
del tree_dict_mcmc, tree_names_mcmc, tree_wts_mcmc
```

```python
# set up the model
model = VBPI(taxa, rootsplit_supp_dict, subsplit_supp_dict, data, pden=np.ones(4)/4, subModel=('JC', 1),
             emp_tree_freq=None, psp=True)
```

```python
# variational parameters
for param in model.parameters():
    print(param.dtype, param.size())
```

```python
# run vbpi.
# the trained model will be saved to 'results/flu100/mcmc_vimco_10_psp.pt'
model.learn(0.001, maxiter=400000, n_particles=10, warm_start_interval=100000, checkpoint_freq=100000, method='vimco',
            save_to_path='../results/flu100/mcmc_vimco_10_psp.pt')
```

```python
# load the ground truth
tree_dict_ground_truth, tree_names_ground_truth, tree_wts_ground_truth = summary('flu100', 'data/flu_data/', samp_size=75001)
emp_tree_freq = {tree_dict_ground_truth[tree_name]:tree_wts_ground_truth[i] for i, tree_name in enumerate(tree_names_ground_truth)}
del tree_dict_ground_truth, tree_names_ground_truth, tree_wts_ground_truth
```

```python
# set up the ground truth in the model
model.emp_tree_freq = emp_tree_freq
trees, emp_freqs = zip(*emp_tree_freq.items())
emp_freqs = np.array(emp_freqs)
model.negDataEnt = np.sum(emp_freqs * np.log(np.maximum(emp_freqs, model.EPS)))
```

```python
# load checkpoints
model.load_from('../results/flu100/mcmc_vimco_10_pspcheckpoint_200000.pt')
```

```python
# compute the KL divergence
model.kl_div()
```

```python
# compute the marginal likelihood
marginal_likelihood_est = np.array([model.lower_bound(n_particles=1000, n_runs=1) for i in range(100)])
```
